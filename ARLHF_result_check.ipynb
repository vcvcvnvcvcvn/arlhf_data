# 初始化

```python
import json
import pandas as pd
import os
```

```python
def result_fetch():
    _ = os.system('wget http://39.98.141.84:3838/iMarxTool/zs_predictions.json')
    _ = os.system('wget http://39.98.141.84:3838/iMarxTool/rm_predictions.json')
    _ = os.system('wget http://39.98.141.84:3838/iMarxTool/sl_predictions.json')
    _ = os.system('wget http://39.98.141.84:3838/iMarxTool/ppo_predictions.json')
    _ = os.system('wget http://39.98.141.84:3838/iMarxTool/ppo_baseline_test_split_metrics.jsonl')
    _ = os.system('wget http://39.98.141.84:3838/iMarxTool/sl_baseline_test_split_metrics.jsonl')
    _ = os.system('wget http://39.98.141.84:3838/iMarxTool/rm_test_split_metrics.jsonl')
```

```python
def json2dict(path):
    with open(path, 'r') as json_file:
        json_list = list(json_file)
    for json_str in json_list:
        result = json.loads(json_str)
    return result['metrics']
```

```python
def show_metric():
    with open('rm_test_split_metrics.jsonl', 'r') as json_file:
        json_list = list(json_file)
    zero_baseline = json.loads(json_list[0])['metrics']
    ppo_bs_result = json2dict('ppo_baseline_test_split_metrics.jsonl')
    sl_bs_result = json2dict('sl_baseline_test_split_metrics.jsonl')
    myrw_result = json2dict('rm_test_split_metrics.jsonl')
    output_df = pd.concat([pd.DataFrame(zero_baseline,index=['zero-shot']).T,pd.DataFrame(ppo_bs_result,index=['ROUGE-PPO']).T,pd.DataFrame(sl_bs_result,index=['supervised learning']).T,pd.DataFrame(myrw_result,index=['RM-PPO']).T],axis = 1)
    return output_df
```

```python
with open('zs_predictions.json', 'r') as json_file:
    zs_txt = json.loads(json_file.readline())
with open('rm_predictions.json', 'r') as json_file:
    rm_txt = json.loads(json_file.readline())
with open('sl_predictions.json', 'r') as json_file:
    sl_txt = json.loads(json_file.readline())
with open('ppo_predictions.json', 'r') as json_file:
    rl_txt = json.loads(json_file.readline())
def show_prediction(idx, show_source = False):
    if show_source:
        print('-'*100)
        print('source: ')
        print(rm_txt[idx]['prompt_text'])
    print('-'*100)
    print('reference: ')
    print(rm_txt[idx]['ref_text'])
    print('-'*100)
    print('reward model-ppo: ')
    print(rm_txt[idx]['generated_text'])
    print('-'*100)
    print('suvervised learning: ')
    print(sl_txt[idx]['generated_text'])
    print('-'*100)
    print('rouge-ppo: ')
    print(rl_txt[idx]['generated_text'])
    print('-'*100)
    print('zero-shot: ')
    print(zs_txt[idx]['generated_text'])
    print('-'*100)
```
